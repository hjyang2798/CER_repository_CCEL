# Bare

## N4

```python

f = open("bader/bare_N4_ACF.dat",'r')
lines = f.readlines()

bader = list()
for i,line in enumerate(lines[2:-4]):
    line = line.split()  # 줄 끝의 줄 바꿈 문자를 제거한다.
    bader.append(line)
f.close()

df_bare_N4 = pd.DataFrame(bader)
df_bare_N4.drop([0],axis=1)
df_bare_N4.columns = ["#","x","y","z","charge","min_dist","atomic vol"]
df_bare_N4 = df_bare_N4.drop(['#','min_dist','atomic vol'],axis=1)

atom_list = ['C'] *20 + ['N'] * 4 + ['Ir'] * 1
df_bare_N4['atom'] = atom_list


electron = [4,5,9]
net_list = list()
for i,atom in enumerate(atom_list) :
    if atom == "C" :
        net_list.append(-float(list(df_bare_N4['charge'])[i]) + electron[0])
    if atom == "N" :
        net_list.append(-float(list(df_bare_N4['charge'])[i]) + electron[1])
    if atom == "Ir" :
        net_list.append(-float(list(df_bare_N4['charge'])[i]) + electron[2])    
        
df_bare_N4['net'] = net_list

df_bare_N4 = df_bare_N4.astype({'x':'float', 'y':'float', 'z':'float', 'charge' : 'float'})

df_bare_N4
```
